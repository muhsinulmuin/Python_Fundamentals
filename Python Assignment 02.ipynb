```python
#1 sum of all numbers in a list 

def sum_of_list_numbers(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(sum_of_list_numbers([1, 2, 3, 4, 5]))      # Test 1 -> 15
print(sum_of_list_numbers([10, 20, 30, 40, 50]))  # Test 2 -> 150
```

```python
#3

def count_greater_than_10(numbers):
    count = 0
    for num in numbers:
        if num > 10:
            count += 1
    return count 
print(count_greater_than_10([12, 10, 9, 21, 21])) 
print(count_greater_than_10([15,70,14,9,2]))
```

```python
#4

def add_and_remove(items, add_item, remove_item):
    print("Original list:", items)
    items.append(add_item)
    print("After adding:", items)
    items.remove(remove_item)
    print("After removing:", items)

add_and_remove([1, 2, 3, 4, 5], 6, 2)
add_and_remove(["a", "b", "c"], "d", "a")
```
