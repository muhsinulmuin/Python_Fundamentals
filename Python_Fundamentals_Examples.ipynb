```python
#prob 15 
# Format the same output string three ways: f-string, .format(), % operator.

product = 'Laptop'
price = 75000.5
discount = 10

# f-string (modern, recommended)
print(f"Product: {product} | Price: {price:.2f} | Discount: {discount}%") 

# .format() method
print("Product: {} | Price: {:.2f} | Discount: {}%".format(product, price, discount))

# % operator (old-style C formatting)
print("Product: %s | Price: %.2f | Discount: %d%%" %(product, price, discount))
```

```python
#prb 17
# area calculator - read length & width, compute area and perimeter.
# tested with 2 different rectangle dimensions. 

def rectangle_calc(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    print(f"Length: {length}, Width: {width}")
    print(f"Area : {area:.2f}")
    print(f"perimeter: {perimeter:.2f}")

#test 1
rectangle_calc(8.5, 4.2)

#test 2
rectangle_calc(12.0, 3.5)
```

```python
#prob 18
# Grade calculator - 5-branch if, elif, else for letter grade, separate Pass/Fail.
# Tested with 2 scores hitting different grade branches.

def grade_calc(score):
    if score >= 90:
        grade = 'A+'
    elif score >= 80:
        grade = 'A'
    elif score >= 60:
        grade = 'B'
    elif score >= 40:
        grade = 'c'
    else:
        grade = 'F'     
    print(f"Score: {score} -> Grade: {grade} | {'Pass' if score >= 40 else 'Fail'}")

#test1 
grade_calc(73)

#test2
grade_calc(32)
```

```python
#prob 19
# Login system - nested if, else checks username first, then password.
# Tested with 3 inputs: correct login, worng password, worng username.

correct_username = "admin"
correct_password = "pass123"


def login(username_input, password_input):
     
     if username_input == correct_username:
         if password_input == correct_password:
             print("Access Granted")
         else:
             print("Wrong Password")
     else:
         print("User Not Found")
        
# Test 1 correct username and password
login("admin", "pass123")

# Test 2 correct username, wrong password
login("admin", "wrongpass")

# Test 3 wrong username entirely
login("guest", "pass123")
    
        
```
