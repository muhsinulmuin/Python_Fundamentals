```python
#prob 18

score = 73

if score >= 90:
    grade = 'A+'
elif score >= 80:
    grade = 'A'
elif score >= 60:
    grade = 'B'
elif score >= '40':
    grade = 'c'
else:
    grade = 'F'

print(f"Grade: {grade}")
print("Pass" if score >= 40 else "Fail")
```
